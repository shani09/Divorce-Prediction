```python
def plot_couples(df, n_samples):  
    
    plt.figure(figsize=(15, 180))
    plt.subplots_adjust(hspace=0.5)

    sam = df.sample(n_samples)

    for i in range(1, n_samples+1):
        # add a new subplot iteratively
        ax = plt.subplot(80, 2, i)

        # gets values of the current couple
        df_row = sam.iloc[i-1].T.to_frame()
        df_row.reset_index(inplace=True)
        couple_ind = df_row.columns[1] # index from original df
        df_row.columns = ['statements', 'score']

        # set background color based on whether divorced or not
        if df_row.iloc[54].score == 0:
            ax.set_facecolor((0.80000, 1.00000, 0.80000, 0.5))
        else:
            ax.set_facecolor((1.00000, 0.80000, 0.80000, 0.5))

        ax.title.set_text(f'couple number {couple_ind}')

        # set the values for y axis
        custom_ylim = (-1, 5)
        plt.setp(ax, ylim=custom_ylim)
        
        # plot positive
        x = range(29)
        y = df_row.iloc[:29].score
        ax.scatter(x, y, c='green')

        # plot negative
        x = list(range(29, 54))
        y = df_row.iloc[29:].score[:-1]
        ax.scatter(x, y, c='red')
```

The function plot_couples defined above recieves a dataframe and a number of couples we want to randomly select from the dataframe and plot.

#### K-Means

I will use the elbow method on K-means to determine the value of K.

In the elbow method we run the K-means algorithm on different values of K (number of clusters), and each time plot the WCSS. WCSS is the sum of squared distance between each point and the centroid in a cluster. When we plot the WCSS with the K value, the plot looks like an elbow. As the number of clusters increases, the WCSS value will start to decrease. The the optimum value for K is the K for which after it the slope of the graph starts to be flat.

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, n_init='auto', random_state =42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(6, 6))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

By looking at the graph, k should be 2 or 3.

I will evaluate KMeans for different values of k using the silhouette score, and compare the results with the result of the elbow method.

```python
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
```

```python
def eval_clustering(model_name):
    
    range_n_clusters = [2, 3, 4, 5, 6]

    for n_clusters in range_n_clusters:
        fig, ax1 = plt.subplots(1, 1)
        fig.set_size_inches(7, 5)

        # The 1st subplot is the silhouette plot
        # The silhouette coefficient can range from -1, 1 but in this example all
        # lie within [-0.1, 1]
        ax1.set_xlim([-0.1, 1])
        # The (n_clusters+1)*10 is for inserting blank space between silhouette
        # plots of individual clusters, to demarcate them clearly.
        ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

        # Initialize the clusterer with n_clusters value and a random generator
        # seed for reproducibility.
        if model_name == 'KMeans':
            clusterer = KMeans(n_clusters=n_clusters, n_init="auto", random_state=42)
        elif model_name == 'Spectral':
            clusterer = SpectralClustering(n_clusters=n_clusters, assign_labels='cluster_qr', eigen_tol=0.0001, random_state=42)
        
        
        # makes clusters and gets the cluster labels
        cluster_labels = clusterer.fit_predict(X)

        # The silhouette_score gives the average value for all the samples.
        # This gives a perspective into the density and separation of the formed
        # clusters
        silhouette_avg = silhouette_score(X, cluster_labels)
        print(
            "For n_clusters =",
            n_clusters,
            "The average silhouette_score is :",
            silhouette_avg,
        )

        # Compute the silhouette scores for each sample
        sample_silhouette_values = silhouette_samples(X, cluster_labels)

        y_lower = 10
        for i in range(n_clusters):
            # Aggregate the silhouette scores for samples belonging to
            # cluster i, and sort them
            ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

            ith_cluster_silhouette_values.sort()

            size_cluster_i = ith_cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_clusters)
            ax1.fill_betweenx(
                np.arange(y_lower, y_upper),
                0,
                ith_cluster_silhouette_values,
                facecolor=color,
                edgecolor=color,
                alpha=0.7,
            )

            # Label the silhouette plots with their cluster numbers at the middle
            ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

            # Compute the new y_lower for next plot
            y_lower = y_upper + 10  # 10 for the 0 samples

        ax1.set_title("The silhouette plot for the various clusters.")
        ax1.set_xlabel("The silhouette coefficient values")
        ax1.set_ylabel("Cluster label")

        # The vertical line for average silhouette score of all the values
        ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

        ax1.set_yticks([])  # Clear the yaxis labels / ticks
        ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
```

```python
eval_clustering('KMeans')
```

I made the following two functions so that I could easily look into the clustering results and evaluate them statistically but also manually.

The first function 'make_clusts_dfs' makes a new dataframe for each cluster, with only the samples that belong to that cluster. It saves the new dataframes in a list and returns it.

The second function 'print_cluster_stats' prints basic stats about a specifies cluster. It prints how many samples/couples are in the cluster, and how many of them got divorced and how many of them didn't get divorced.

```python
def make_clusts_dfs(df_clusts, n_clusts):
    clusts_dfs = []
    for i in range(n_clusts):
        clusts_dfs.append(df_clusts[df_clusts.cluster == i].copy())
        clusts_dfs[i].drop(columns = ['cluster'], inplace=True)
    return clusts_dfs
```

```python
def print_cluster_stats(df, cluster_num):
    size = df.shape[0]
    n_divorced = df[df.Divorced == 0].shape[0]
    n_not_divorced = df[df.Divorced == 1].shape[0]
    print(f'Cluster number {cluster_num} has {size} couples, of them {n_divorced} are divorced, and {n_not_divorced} are not divorced.')
```

### PCA Dimention Reduction

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

PCA is affected by scale, so we need to scale the features in the data before applying PCA.

```python
# scaling the data for PCA
scaler=StandardScaler()
scaler.fit(X)
scaled_data=scaler.transform(X)
```

```python
# perform PCA
pca = PCA()
pca.fit(scaled_data) 
pca_data = pca.transform(scaled_data) # get PCA coordinates for scaled_data
```

```python
# Draw a scree plot and a PCA plot
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
labels_short = [str(x) for x in range(1, len(per_var)+1)]

plt.figure(figsize=(10, 5))
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels_short)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

We can see that almost all of the variation (above 70%) is along the first principal component, so a graph that plots PC1 and PC2 should do a good job representing the original data.

```python
# makes the plot using PC1 and PC2
pca_df = pd.DataFrame(pca_data, index=X.index, columns=labels)

plt.figure(figsize=(10, 5))
plt.scatter(pca_df.PC1, pca_df.PC2, c=y, cmap='RdYlGn_r')
# plt.scatter(pca_df.PC1, pca_df.PC2, c=y)
plt.title('My PCA Graph')
plt.xlabel('PC1 - {0}%'.format(per_var[0]))
plt.ylabel('PC2 - {0}%'.format(per_var[1]))
 
for sample in pca_df.index:
    plt.annotate(sample, (pca_df.PC1.loc[sample], pca_df.PC2.loc[sample]))
```

On the left we can see a clear cluster of (mostly) dark green dots, and on the right a scatter of the rest of the samples, not in a clear cluster.\
The separation between them is horizontal, meaning that it can be attributed to PC1, or the features included in PC1.

On the right of the PCA graph we can see a fairly sparse cluster compared to the cluster on the left. On the right of this cluster we see a light-green and green "sub-cluster", and on the upper-left a sparse red "sub-cluster".

Since most of the samples are green (couples that don't feel like strangers in the same environment), we would like to figure out what differentiates the green samples from the red ones. 

We notice that PC2 doesn't seperable the green and red dots very well inside the right cluster, since there isn't a clear gap between them. In addition, there are orange and light green dots near the red dots, not only dark green. Thus, I will call the dark green dots in the left cluster couples in the "safe zone" - couples that are clearly far from the red, orange, and even light green dots. Similarly, I will call couples in the right cluster couples in the "danger zone" - couples that there is very little (3.8%) to explain why they are orange, light-green, or dark-green as opposed to red.

We would like to look into what features make up the first principal component (PC1), in order to understand which features explain most the seperation between couples in the "safe zone" and couples in the "danger zone".

In PCA , the component loadings are analogous to correlation coefficients, squaring them give the amount of explained variation. Therefore the component loadings tell us how much of the variation in a variable is explained by the component. I will first remove features with correlations greater that 0.95 in absolute value, then continue to remove features with a lower correlations that are still high.

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = []  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname1 = corr_matrix.columns[i]  # getting the name of column 1
                colname2 = corr_matrix.columns[j]  # getting the name of column 2
                if (colname1, colname2) not in col_corr and (colname2, colname1) not in col_corr:
                    col_corr.append((colname1, colname2))
    return col_corr
```

#### K-Nearest-Neighbors

Based the insights gained from the project up until now, and based on the assumption that similar inputs produce similar outputs, I will start by implementing a K-Nearest-Neighbors model. 

```python
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier, KNeighborsRegressor
```

I will start by looking the results for KNN with different values of k (with the default parameter 'weights': 'uniform').

Since that class sizes are unbalanced, I used the 'balanced_accuracy' scoring metric, which is the average of recall obtained on each class. Cross_val_score used stratified cross validation (balanced the ensures that each fold has the same proportion of observations from each class), which is desired for imbalanced data. Due to the fact that there are few samples in classes 1 and 2, the maximum number of folds that allow this is 3.

```python
# search for an optimal value of K for KNN
k_scores = []
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=3, 
                             scoring='balanced_accuracy')
    k_scores.append(scores.mean())

plt.figure(figsize=(8, 4))
plt.plot(range(1,21), k_scores)
plt.title('KNN - balanced accuracy vs. K')
plt.xlabel('K')
plt.ylabel('CV balanced accuracy')
```

We can see that the balanced accuracy is a bit higher (though still not very high) with k=1. For k >= 2 the balanced accuracy is constant at below 0.34.\
These results are probably there weren't many samples to begin with, even less samples after splitting to training and testing sets, and even less in each fold of the cross validation.

```python
# define transforms to apply to the tabular dataset
procs = [Categorify]
# define the dependent variable (y value)
dep_var = 'Strangers'
# define columns that are continuous / categorical
cont,cat = cont_cat_split(df2, dep_var=dep_var) 
```

```python
# define TabularDataLoaders object

dls=TabularDataLoaders.from_df(df2, procs=procs, cat_names= cat, cont_names = cont, y_names = dep_var, y_block = CategoryBlock, bs=60)
```

```python
# use show_batch() to see a sample batch including x and y values
dls.show_batch()
```

```python
# redefined F1Score from fastai to include ignore zero_division warning.

from sklearn.metrics import f1_score

def F1Score(axis=-1, labels=None, pos_label=1, average='binary', sample_weight=None):
    "F1 score for single-label classification problems"
    return skm_to_fastai(f1_score, axis=axis,
                         labels=labels, pos_label=pos_label, average=average, sample_weight=sample_weight,
                        zero_division=0)
```

As explained previously, because the dataset is imbalanced the accuracy is not a good metric in this case. Instead we will use the f1 score as a metric.

After many tries and variations of the sizes of the neural network layers we get the following result:

```python
learn = tabular_learner(dls, layers=[80, 50, 20, 3], metrics=F1Score(labels=[0,1,2], average='weighted'))
plt.figure(figsize=(10, 5))
with learn.no_bar(), learn.no_logging():
    learn.fit(400)
    learn.recorder.plot_loss()
```

```python
learning_interpreter = ClassificationInterpretation.from_learner(learn)
learning_interpreter.plot_confusion_matrix()
```

Our neural network perfectly classified all of the samples in classes 0 and 1. It correctly classified 2 out of the 3 samples in class 2.
